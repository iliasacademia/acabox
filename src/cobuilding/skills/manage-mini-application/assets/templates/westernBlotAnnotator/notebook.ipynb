```
params_json = '{"font_size":14,"input_image":"","script_path":".applications/westernBlotAnnotator/scripts/analyze.py","outdir":".applications/westernBlotAnnotator/output"}'
```

```
import json, importlib.util

params = json.loads(params_json)

# Load analyze.py as a module from this app's scripts/ directory.
spec = importlib.util.spec_from_file_location('analyze', params['script_path'])
analyze = importlib.util.module_from_spec(spec)
spec.loader.exec_module(analyze)

# run() writes run_metadata.json into outdir and returns the result dict.
analyze.run(params['input_image'], params['outdir'])
```
