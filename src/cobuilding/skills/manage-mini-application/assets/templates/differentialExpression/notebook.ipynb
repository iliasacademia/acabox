# Differential Expression Analysis

This notebook backs the mini-app `differentialExpression`.

Edit params via the app UI, or directly in the parameters cell below.
Input files live under `./input/<slot>/` (inside this directory).
Paths in `params_json` are workspace-relative
(`.applications/differentialExpression/...`), which is what the
cobuild kernel and React app expect.


```r
params_json <- '{
  "alpha": 0.05,
  "coldata_file": "",
  "counts_file": "",
  "denominator_level": "",
  "design_formula": "",
  "design_mode": "variable",
  "design_variable": "",
  "lfc_threshold": 1,
  "min_count": 10,
  "min_samples": 3,
  "numerator_level": "",
  "orgdb": "org.Hs.eg.db",
  "shrink": false
}'
```

```r
source(".claude/skills/differential-expression/scripts/differential_expression.R")

params <- jsonlite::fromJSON(params_json)

# outdir is supplied by the app at run time; default for stand-alone
# notebook execution is the app's output/ directory.
outdir <- if (!is.null(params$outdir)) params$outdir else "./output"

results <- run_differential_expression(
  counts_file = params$counts_file,
  coldata_file = params$coldata_file,
  outdir = outdir,
  design_variable = params$design_variable,
  design_formula = params$design_formula,
  denominator_level = params$denominator_level,
  numerator_level = params$numerator_level,
  min_count = params$min_count,
  min_samples = params$min_samples,
  alpha = params$alpha,
  lfc_threshold = params$lfc_threshold,
  shrink = params$shrink,
  orgdb = params$orgdb
)

result_info <- list(
  status = "success",
  files = list(
    volcano_plot = file.path(outdir, "volcano_plot.csv"),
    ma_plot = file.path(outdir, "MA_plot.csv")
  )
)
jsonlite::write_json(result_info, file.path(outdir, "results.json"), auto_unbox = TRUE)
```
